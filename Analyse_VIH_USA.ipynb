```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plot
# Pour afficher le tableau entirement
pd.set_option('display.max_columns', None) 
pd.set_option('display.max_rows', None)
pd.set_option('display.max_colwidth', None) 

# On rcupre le fichier CSV et on drop les columns inutiles
file = pd.read_csv("./Big_Cities_Health_Data_Inventory.csv")
file = file.drop(['Notes', 'Methods', 'Source', 'BCHC Requested Methodology'], axis=1)

# On drop les lignes dont une valeur est nulle
file = file.dropna()

# On drop les lignes dont les columns ont une certaine value : 
def filter_rows_by_values(df, col, values):
    return df[df[col].isin(values)]


# On choisi 1 seul indicateur
file = filter_rows_by_values(file, "Indicator Category", ["HIV/AIDS"])
file1 = filter_rows_by_values(file, "Indicator", ["Persons Living with HIV/AIDS Rate (Per 100,000 people)"])
file2 = filter_rows_by_values(file, "Indicator", ["HIV-Related Mortality Rate (Age-Adjusted; Per 100,000 people)"])
file3 = filter_rows_by_values(file, "Indicator", ["HIV Diagnoses Rate (Per 100,000 people)"])
file4 = filter_rows_by_values(file, "Indicator", ["AIDS Diagnoses Rate (Per 100,000 people)"])

# On affiche le tableau entirement
file.head(len(file.index))
```

```python
# On fait un groupby sur les annes afin d'avoir une variable contenant uniquement 1 indicateur, 
# Evolution de Persons Living with HIV/AIDS Rate (Per 100,000 people) par anne

file1 = file1.groupby(["Year"]).Value.mean()
print(file1)
file1.plot.bar()
```

```python
# On fait un groupby sur les annes afin d'avoir une variable contenant uniquement 1 indicateur, 
# Evolution de HIV-Related Mortality Rate (Age-Adjusted; Per 100,000 people) par anne

file2 = file2.groupby(["Year"]).Value.mean()
print(file2)
file2.plot.bar()
```

```python
# On fait un groupby sur les annes afin d'avoir une variable contenant uniquement 1 indicateur, 
# Evolution de HIV Diagnoses Rate (Per 100,000 people) par anne

file3 = file3.groupby(["Year"]).Value.mean()
print(file3)
file3.plot.bar()
```

```python
# On fait un groupby sur les annes afin d'avoir une variable contenant uniquement 1 indicateur, 
# Evolution de AIDS Diagnoses Rate (Per 100,000 people) par anne

file4 = file4.groupby(["Year"]).Value.mean()
print(file4)
file4.plot.bar()
```

```python
# Data
df=pd.DataFrame({'x_values': range(1,6), 'y1_values': file4, 'y2_values': file2, 'y3_values': file3 })
 
# multiple line plots
plot.plot( 'x_values', 'y1_values', data=df, marker='o', markerfacecolor='blue', markersize=12, color='skyblue', linewidth=4, label="AIDS Diagnoses Rate")
plot.plot( 'x_values', 'y2_values', data=df, marker='o', markerfacecolor='red', markersize=12, color='skyblue', linewidth=4, label="HIV-Related Mortality Rate")
plot.plot( 'x_values', 'y3_values', data=df, marker='o', markerfacecolor='green', markersize=12, color='skyblue', linewidth=4, label="HIV Diagnoses Rate")

# show legend
plot.legend()

# show graph
```
